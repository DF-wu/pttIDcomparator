```python
# 用比較快快的json parser : ujson
try:
    import ujson as json
except ImportError:
    try:
        import simplejson as json
    except ImportError:
        import json

import math


# load tfidf
def readJson2Data(filename):
  f = open(filename, "r", encoding='UTF-8')
  docText = f.read()
  f.close()
  return json.loads(docText)
```

```python

SAMPLE1000 = 'sample1000_gossiping.json'
FILE_NAME = 'gossiping_dataset_seg_filtered.json'
data = readJson2Data(FILE_NAME)

```

```python
# all author dict
docID_name = {}
for i in data:
    try:
        name = i['author']
        name = name.split( )
        usrID = name[0]
        docID_name[usrID] = list()
        
    except:
        print(i['author'])
```

```python

#  map<author,[docID...]>

# 把文章id拿出來 歸類到同作者下
for doc in data:
    # 有的名子有問題
    try:
        usrID = doc['author'].split( )
        usrID = usrID[0]
        docID_name[usrID].append(doc['article_id'])
    except:
        print(doc['author'] + doc['article_id'] )
```

```python
# tools
def saveSmallData(d):
    jsObj = json.dumps(d)
    
    fileObject = open('data_user-docIDs.json', 'w')
    fileObject.write(jsObj)
    fileObject.close()  
saveSmallData(docID_name)
```
