```python
import json
import math
```

```python
# docfile = "ReutersCorn-sample10.json"
docfile = "ReutersCorn-train.json"
```

```python
f = open(docfile, "r", encoding='UTF-8')
dataset = json.loads(f.read())
f.close()
```

```python
import nltk
nltk.download('punkt')

from nltk.stem.porter import *
stemmer = PorterStemmer() # 未來建議各位使用 SnowballStemmer
```

```python
wordListDict = dict()
docWordDict = dict()

for data in dataset:
    tokens = nltk.word_tokenize(data["text"])

    # sum up tf in cur doc
    curDocDict = dict()
    for tok in tokens:
        wordIndex = stemmer.stem(tok.lower())
        if wordIndex not in curDocDict:
            curDocDict[wordIndex] = 0
        curDocDict[wordIndex] += 1
    
    # create doc word dict
    docWordDict[data["docID"]] = curDocDict
    
    # create inverted list
    for word in curDocDict:
        if word not in wordListDict:
            wordListDict[word] = []
        if len(wordListDict[word]) == 0 or wordListDict[word][len(wordListDict[word]) - 1]["docID"] != data["docID"]:
            wordListDict[word].append({
                "docID": data["docID"],
                "tf": curDocDict[word]
            })
```

```python
for docID in docWordDict:
    for word in docWordDict[docID]:
        tf = docWordDict[docID][word]
        df = len(wordListDict[word])
        N = len(docWordDict)
        score = (1 + math.log10(tf)) * math.log10(N / df)
        docWordDict[docID][word] = score
```

```python
normDocWordDict = dict()

for docID in docWordDict:
    normDocWordDict[docID] = dict()
    totalTf = 0
    for word in docWordDict[docID]:
        totalTf += math.pow(docWordDict[docID][word], 2)

    base = math.sqrt(totalTf)
    for word in docWordDict[docID]:
        normDocWordDict[docID][word] = docWordDict[docID][word] / base
```

```python
def cosSim(id1, id2):
    score = 0
    for word in normDocWordDict[id1]:
        if word in normDocWordDict[id2]:
            score += normDocWordDict[id1][word] * normDocWordDict[id2][word]
    return score

def formatOutput(score):
    return "{:.5f}".format(score)

def calcDf(word):
    return len(wordListDict[word])

def printInvertList(word):
    invList = wordListDict[word]
    result = "{}, df = {},".format(word, calcDf(word))
    for obj in invList:
        result += "({},{})".format(obj["docID"], obj["tf"])
    return result
```
